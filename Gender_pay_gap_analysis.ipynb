```python
# Read dataset
filename = 'Gender pay gap data - Sheet5 (4).csv'
df = pd.read_csv(filename)
```

```python
df.sample(5)
```

```python
# Check for datatypes
df.dtypes
```

```python
# Remove commas from columns 'FT weekly', 'Male FT weekly', 'Female FT weekly'
# 'AGP FT', 'AGP male FT' and 'AGP female FT'
df['FT weekly'] = df['FT weekly'].str.replace(',', '')
df['Male FT weekly'] = df['Male FT weekly'].str.replace(',', '')
df['Female FT weekly'] = df['Female FT weekly'].str.replace(',', '')
df['AGP FT'] = df['AGP FT'].str.replace(',', '')
df['AGP male FT'] = df['AGP male FT'].str.replace(',', '')
df['AGP female FT'] = df['AGP female FT'].str.replace(',', '')
```

```python
# Check that changes have been implemented
df.sample(5)
```

```python
# Change the datatype of the columns from object to numeric value
df['Male FT weekly'] = pd.to_numeric(df['Male FT weekly'])
df['FT weekly'] = pd.to_numeric(df['FT weekly'])
df['Female FT weekly'] = pd.to_numeric(df['Female FT weekly'])
df['AGP FT'] = pd.to_numeric(df['AGP FT'])
df['AGP male FT'] = pd.to_numeric(df['AGP male FT'])
df['AGP female FT'] = pd.to_numeric(df['AGP female FT'])
```

```python
df.dtypes
```

```python
# Check for null values 
df.isnull().sum()
```

```python
# Check the shape of dataset
df.shape
```

```python
# Get a summary of the data
df.describe()
```

```python
# Rename columns

df.rename(columns={'Male FT weekly':'Male_Weekly'}, inplace=True)
df.rename(columns={'Female FT weekly':'Female_Weekly'}, inplace=True)
df.rename(columns={'AGP male FT':'Male_AGP'}, inplace=True)
df.rename(columns={'AGP female FT':'Female_AGP'}, inplace=True)
df.rename(columns={'Gender pay gap (%)':'Gender_pay_gap'}, inplace=True)
```
